# Web Scraping to Populate SQL Database

```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
from datetime import datetime
from pandas.tseries.offsets import DateOffset
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)

sym = input('Input stock ticker ')
url = 'http://openinsider.com/screener?s='+sym+'&o=&pl=&ph=&ll=&lh=&fd=730&fdr=&td=0&tdr=&fdlyl=&fdlyh=&daysago=&xp=1&xs=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=0&cnt=100&page=1'
      
# figure out how to make the requests automated, and scraping the URL while passing in certain companies. 
```

```python
df = pd.read_html(url, skiprows = 1, header = 0)
```

```python
df = df[6]
df.drop(['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15' ], inplace = True, axis = 1)
row_zero = df.columns
```

```python
column_names = ['Type','FilingDate','TradeDate','Ticker','InsiderName', 'Title', 'TradeType', 'Price','Qty','Owned',
               'OwnPctChange', 'Value']

df.columns = column_names

df["Type"] = df['Type'].fillna('S')
```

```python
data = []


data.insert(0,{'Type': row_zero[0] ,'FilingDate' : row_zero[1] ,'TradeDate' : row_zero[2],'Ticker' : row_zero[3] ,'InsiderName' : row_zero[4], 'Title' : row_zero[5], 'TradeType' : row_zero[6], 'Price' : row_zero[7],'Qty' : row_zero[8],'Owned' : row_zero[9],'OwnPctChange' : row_zero[10], 'Value': row_zero[11]})

p = pd.DataFrame(data)

p['Qty'] = p['Qty'].str.replace(',', '').astype(int)
p['Owned'] = p['Owned'].str.replace(',', '').astype(int)
```

```python
df = pd.concat([p, df], ignore_index=True)
df
```

```python
# clean up dateframe types, drop symbols that affect floats and integers. 

df['Price'] = df['Price'].str.replace('$', '').str.replace(',', '').astype(float)
df['Value'] = df['Value'].str.replace('$', '').str.replace(',', '').astype(int)


df['OwnPctChange'] = df['OwnPctChange'].str.replace('%', '').astype(int)
df['FilingDate'] = pd.to_datetime(df['FilingDate'])
df['TradeDate'] = pd.to_datetime(df['TradeDate'])
df[['Type', 'Ticker', 'InsiderName', 'Title', 'Trade-Type']] = df[['Type', 'Ticker', 'InsiderName', 'Title', 'TradeType']].astype(str)
df = df.drop('Trade-Type', 1)
```

```python
insiders = df
insiders
```

# Write Stock and Insider Data to Database

```python
import sqlalchemy
password = os.environ.get('LAPTOP_PASS')

sym = sym.upper()
stock = pd.read_csv(sym+'_df.csv')

engine = sqlalchemy.create_engine('mysql+pymysql://root:'+str(password)+'@localhost:3306/Stock_DB')
engine
```

```python
insiders.to_sql(
    name = 'insiders',
    con = engine,
    index = False,
    if_exists = 'replace'
)
```

```python
stock.to_sql(
    name = 'stock',
    con = engine,
    index = False,
    if_exists = 'replace'
)
```
